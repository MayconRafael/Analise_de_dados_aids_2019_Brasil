```python
import matplotlib
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
low_memory=False,
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
df = pd.read_excel(r'C:\Users\User\Dados-AIDS-Genero-Escolaridade.xlsx')
```

```python
df.info()
```

```python
x = np.arange(len(df.Escolaridade))
width = 0.20
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, (df.Masculino), width, label='Masculino')
rects2 = ax.bar(x + width/2, (df.Feminino), width, label='Feminino')

ax.set_ylabel('Casos')
ax.set_title('Casos de AIDS correlacionados ao Ensino')
ax.set_xticks(x)
ax.set_xticklabels(df.Escolaridade)
ax.legend()

plt.show()
plt.rcParams["figure.figsize"] = [20, 5]
```
